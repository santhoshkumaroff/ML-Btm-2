```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df_sd = pd.read_csv('standardscaler.csv')
```

```python
df_sd
```

```python
df_sd.info()
```

```python
X = df_sd.drop('Target',axis=1)
y = df_sd['Target']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y,train_size=0.8,random_state=42)
```

## Without Using StandardScaler()

```python
#model_creation
model = LinearRegression()
```

```python
#fit /train the model
model.fit(xtrain,ytrain)
```

```python
test_score = model.score(xtest,ytest)
print(f"Test score : {test_score}")
```

```python
train_score = model.score(xtrain,ytrain)
print(f"Train score : {train_score}")
```

## With Using StandardScaler()

```python
#creating object for StandardScaler() to access fit_transform,transform

scaler = StandardScaler()
```

```python

xtrain_scaled = scaler.fit_transform(xtrain) #learn and tranform the values for the model training
xtest_scaled = scaler.transform(xtest) #just transform the values for model testing
```

```python
model = LinearRegression() #model creation
```

```python
model.fit(xtrain_scaled,ytrain) #model fit/train
```

```python
#evaluation for test
test_score = model.score(xtest_scaled,ytest)
print(f"Test score : {test_score}")
```

```python
#evaluation for train
train_score = model.score(xtrain_scaled,ytrain)
print(f"Train score : {train_score}")
```
